# Исследование рынка компьютерных игр 

**Описание проекта:**
Интернет-магазин «Стримчик» продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Доступны данные до 2016 года. Представим, что сейчас декабрь 2016 г., и вы планируете кампанию на 2017-й. 

**Цель работы:** 
Выявить определяющие успешность игры закономерности. 
Отработать принцип работы с данными для прогнозирования продаж. 
Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

**Источник данных:**
файл с выгрузками:
- /datasets/games.csv - данные из открытых источников;

**Описание данных:**

- Name — название игры
- Platform — платформа
- Year_of_Release — год выпуска
- Genre — жанр игры
- NA_sales — продажи в Северной Америке (миллионы проданных копий)
- EU_sales — продажи в Европе (миллионы проданных копий)
- JP_sales — продажи в Японии (миллионы проданных копий)
- Other_sales — продажи в других странах (миллионы проданных копий)
- Critic_Score — оценка критиков (максимум 100)
- User_Score — оценка пользователей (максимум 10)
- Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

Данные за 2016 год могут быть неполными.

**План работы**:
- Обзор данных и предобработка данных.
- Исследовательский анализ данных.
- Портрет пользователя каждого региона.
- Проверка гипотез.
- Общий вывод.

```python
# Функция для вывода значений на графиках seaborn
def show_values(axs, orient="v", space=.01):
    def _single(ax):
        if orient == "v":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() / 2
                _y = p.get_y() + p.get_height() + (p.get_height()*0.01)
                value = '{:.1f}'.format(p.get_height())
                ax.text(_x, _y, value, ha="center") 
        elif orient == "h":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() + float(space)
                _y = p.get_y() + p.get_height() - (p.get_height()*0.5)
                value = '{:.1f}'.format(p.get_width())
                ax.text(_x, _y, value, ha="left")

    if isinstance(axs, np.ndarray):
        for idx, ax in np.ndenumerate(axs):
            _single(ax)
    else:
        _single(axs)
```

```python
# check
# пропущенные значения бары

def pass_value_barh(dfg):
    try:
        (
            (dfg.isna().mean()*100)
            .to_frame()
            .rename(columns = {0:'percent_missing_values'})
            .query('percent_missing_values>0 ')
            .sort_values(by = 'percent_missing_values', ascending = True)
            .plot(kind = 'barh', figsize = (19,6), legend = True, fontsize = 16)
         
        );    
    except:
        print('пропусков не осталось :) или произошла ошибка в первой части функции ')
    plt.title('Пропущенные значения, %' + "\n", fontsize=22, color='SteelBlue')
    plt.xlabel('Процент пропусков', fontsize=16)
    plt.ylabel('Столбцы с пропусками', fontsize=16)
    plt.xlim(0, 100)
```

#### Влияние на продажи отзывов пользователей и критиков

Посмотрим, как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков.  Построим диаграмму рассеивания и посчитаем корреляцию между отзывами и продажами.

```python
# Построим диаграмму рассеивания и посчитаем корреляцию (по платформе PS4):
response_ru = ['критики','пользователи' ]
response=['critic_score','user_score']

# Получаем фигуру и оси
fig, ax = plt.subplots(nrows=1,ncols=2, sharey=True, figsize=(20, 8))
fig.suptitle('Зависимость продаж от отзывов', fontsize=18, fontweight='bold')

# Построим диаграмму рассеивания
for i in range(2):
        report = df_actual[df_actual['platform']=='PS4']
        report.plot(x=response[i],
                    y='total_sales',
                     kind='scatter',
                     grid='True',
                     color='blue',
                     ax=ax[i],
                     title='Платформа PS4 - '+ response_ru[i],
                     xlabel='Отзывы - '+ response_ru[i],
                     ylabel='Продажи')
        print('Корреляция продаж по PS4 с отзывами - ' + response_ru[i]+':',
                   round(df_actual[response[i]].corr(df_actual['total_sales'][df_actual['platform']=='PS4']),2))
plt.show()     
```

Соотнесем выводы с продажами игр на других платформах из топ 3 -  XOne, 3DS.

```python
# Построим диаграмму рассеивания и посчитаем корреляцию:
for platform in ('XOne', '3DS','PS3'):
        # Получаем фигуру и оси
        fig, ax = plt.subplots(nrows=1,ncols=2, sharey=True, figsize=(20, 8))
        fig.suptitle('Зависимость продаж от отзывов - '+platform, fontsize=18, fontweight='bold')

        # Построим диаграмму рассеивания
        for i in range(2):
                report = df_actual[df_actual['platform']==platform]
                report.plot(x=response[i],
                            y='total_sales',
                             kind='scatter',
                             grid='True',
                             color='blue',
                             ax=ax[i],
                             title='Платформа '+platform+' - '+ response_ru[i],
                             xlabel='Отзывы - '+ response_ru[i],
                             ylabel='Продажи')
                print('Корреляция продаж по платформе '+platform+' с отзывами - ' + response_ru[i]+':',
                           round(df_actual[response[i]].corr(df_actual['total_sales'][df_actual['platform']==platform]),2))
        plt.show()  
```

### Проверим гипотезы

Для проверки всех гипотез используем специальный метод для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам: scipy.stats.ttest_ind (array1, array2, equal_var).

#### Гипотеза о различии средних пользовательских рейтингов платформ Xbox One и PC

- Н0: Средние пользовательские рейтинги платформ Xbox One и PC равны.
- H1: Средние пользовательские рейтинги платформ Xbox One и PC различаются.

```python
# выделим выборки для проверки гипотез
platform_xone = df_actual[(df_actual['platform']=='XOne')\
                           &(df_actual['year_of_release']>=2013)]['user_score'].dropna()
platform_pc = df_actual[(df_actual['platform']=='PC')\
                         &(df_actual['year_of_release']>=2013)]['user_score'].dropna()

print('Размер выборки - ', platform_xone.shape,
      'Размер выборки - ', platform_pc.shape)
print('Средняя оценка XOne - ', round(platform_xone.mean(),1),
      'Средняя оценка PC - ', round(platform_pc.mean(),1))
```

В итоге, у нас достаточно большие (более 30) и равные по размеру выборки и нет оснований думать, что дисперсии генеральных совокупностей разные. Будем сравнивать средние t-тестом, не указывая параметр equal_var: пусть он примет значение True.

```python
# Выполним проверку гипотезы методом ttest_ind

alpha = 0.01 # уровень статистической значимости
# если p-value окажется меньше него, отвергнем гипотезу

results = st.ttest_ind(platform_xone, platform_pc)

print(f'Значение p-value {"{0:.3f}".format(results.pvalue)}')

if results.pvalue < alpha:
    print(f'Отвергаем нулевую гипотезу. Разница статистически значима.')
else:
    print(f'Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя.')
    
```

Значение p-value,равное 0.105, говорит о том, что с вероятностью в 10,5% такое или большее различие между средними оценками платформ можно получить случайно. Эта вероятность выше выбранного уровня статической значимости и сделать вывод о значимом различии между средними оценками платформы XOne и PC нельзя.

```python
# Приведены два датасета: сумма покупок, совершённых за месяц посетителями ...

sample_1 = [3071, 3636, 3454, 3151, 2185, 3259, 1727, 2263, 2015,
2582, 4815, 633, 3186, 887, 2028, 3589, 2564, 1422, 1785,
3180, 1770, 2716, 2546, 1848, 4644, 3134, 475, 2686,
1838, 3352]
sample_2 = [1211, 1228, 2157, 3699, 600, 1898, 1688, 1420, 5048, 3007,
509, 3777, 5583, 3949, 121, 1674, 4300, 1338, 3066,
3562, 1010, 2311, 462, 863, 2021, 528, 1849, 255,
1740, 2596]
alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу
results = st.ttest_ind(
sample_1,
sample_2)
print('p-значение:', results.pvalue)
if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:

    print("Не получилось отвергнуть нулевую гипотезу")
```

#### Гипотеза о различии средних пользовательских рейтингов жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») 

- Н0: Средние пользовательские рейтинги жанров Action и Sports равны.
- H1: Средние пользовательские рейтинги платформ жанров Action и Sports различаются.

```python
 # выделим выборки для проверки гипотез
genre_action = df_actual[(df_actual['genre']=='Action')]['user_score'].dropna()
genre_sports = df_actual[(df_actual['genre']=='Sports')]['user_score'].dropna()

print('Размер выборки - ', genre_action .shape,
      'Размер выборки - ', genre_sports.shape)
print('Средняя оценка action - ', round(genre_action .mean(),2),
      'Средняя оценка sports - ', round(genre_sports.mean(),2))
```

В итоге, у нас достаточно большие (более 30) и разные по размеру выборки и есть основания думать, что дисперсии генеральных совокупностей разные. Будем сравнивать средние t-тестом, указывая параметр equal_var: пусть он примет значение False.

```python
# Выполним проверку гипотезы методом ttest_ind

alpha = 0.01 # уровень статистической значимости
# если p-value окажется меньше него, отвергнем гипотезу

results = st.ttest_ind(genre_action, genre_sports,equal_var = False)

print('p-value:', round(results.pvalue,2))


if results.pvalue < alpha:
    print(f'Отвергаем нулевую гипотезу. Разница статистически значима.')
else:
    print(f'Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя.')
    
```

Значение p-value,равное 0.0000, говорит о том, что вероятность получить такое или большее различие между средними оценками жанров случайно практически равна нулю.
